# Pyber Data Analysis

# Observed trends

* Results of analysis confirm hipotesis that total number of drivers is significantly higher in urban cities, with over 80% of total drivers, compared to suburban (16.5%) and rural (2.6%) city types. Some of possible reasons are convenience, car ownership and high parking fares in the city.

* As shown in bubble plot below, the type of city with the most dispersion in average fares is rural city type (average fares between 25 and 45 dollars) in comparisement to suburban (average fares between 25 and 35 dollars) and urban city type (average fares between 20 and 30 dollars). This might be connected to lower number of drivers, bigger distances between towns and consumers habbits and lifestyle. 

* Rural and suburban city types also have significantly lower number of rides then urban city type, which might suggest people in rural and suburban areas rather use other ways of transportation then shared ride. 

* I would suggest including city population and demographic data into our research, which would give us more accurate insights into reasons behind these results.

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_df, city_df, on="city", how="left")

# Display the data table for preview
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

pyber_grouped = pyber_df.groupby(["type","city"])

ride_share_data = pyber_grouped[["fare"]].mean()

ride_share_data["ride count"] = pyber_grouped[["ride_id"]].count()

ride_share_data["driver count"] = pyber_grouped[["driver_count"]].min()
```

```python
# Build the scatter plots for each city types

colors = ['coral', 'lightskyblue', 'gold']
city_type = list(pyber_df["type"].unique())

for i in range(len(city_type)):
    c = colors[i] 
    t = city_type[i]
       
    plt.scatter(ride_share_data.loc[[t], ["city","ride count"]],
            ride_share_data.loc[[t], ["city","fare"]],  
            facecolors= c, edgecolors="black", alpha=0.75,
            label = t, marker= "o", linewidth=1, 
            s= (ride_share_data.loc[[t], ["driver count"]])*10)
    
# Create a legend

lgnd = plt.legend(title="City Types", loc="upper right",)
                            
lgnd.legendHandles[0]._sizes = [40]
lgnd.legendHandles[1]._sizes = [40]
lgnd.legendHandles[2]._sizes = [40]

#Incorporate the other graph properties

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid(color='grey', linestyle='-', linewidth=0.5,)
plt.ylim(ymin=19, ymax=44)
plt.xlim(xmin=0, xmax=41)

# Incorporate a text label regarding circle size

plt.figtext(0.92, 0.6, 'Note:' + '\n' +'Circle size correlates with driver count per city')

# Save Figure

plt.savefig(".Pyber_Ride_Sharing_Data_(2016).png")

# Show plot

plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents


ride_share_data["fare total"] = pyber_grouped[["fare"]].sum()
ride_share_data_0 = ride_share_data.reset_index("city")

# Build pie chart

city_type = list(pyber_df["type"].unique())
city_type.sort()
city_fare = []

for i in range(len(city_type)):
    city_fare.append(ride_share_data_0.loc[city_type[i],"fare total"].sum())


explode = (0, 0, 0.1)
colors = ['gold', 'lightskyblue', 'lightcoral']
    
plt.pie(city_fare, explode=explode, labels=city_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150, frame=False)

plt.title("% of Total Fares by City Type")
plt.axis("auto")

# Save figure
plt.savefig(".%_of_Total_Drivers_by_City_Type.png")

# Build Pie Chart
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
city_rides = []

for i in range(len(city_type)):
    city_rides.append(ride_share_data_0.loc[city_type[i],"ride count"].sum())

# Build pie chart
explode = (0, 0, 0.1)
colors = ['gold', 'lightskyblue', 'lightcoral']
    
plt.pie(city_rides, explode=explode, labels=city_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150, frame=False)

plt.title("% of Total Rides by City Type")
plt.axis("auto")

# Save image and print the final product to the screen
plt.savefig(".%_of_Total_Rides_by_City_Type.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_drivers = []

for i in range(len(city_type)):
    city_drivers.append(ride_share_data_0.loc[city_type[i],"driver count"].sum())

# Build pie chart
explode = (0, 0, 0.1)
colors = ['gold', 'lightskyblue', 'lightcoral']
    
plt.pie(city_drivers, explode=explode, labels=city_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150, frame=False)

plt.title("% of Total Drivers by City Type")
plt.axis("auto")


# Save image and print the final product to the screen
plt.savefig(".%_of_Total_Drivers_by_City_Type.png")
plt.show()
```
